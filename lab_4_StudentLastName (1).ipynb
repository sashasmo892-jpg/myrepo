## 1. Обчислення факторіалу (цикл for)
Асимптотична складність: **O(n)**

```

def factorial_for(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

factorial_for(5)
```

## 2. Обчислення факторіалу (рекурсія)
Асимптотична складність: **O(n)**

```

def factorial_rec(n):
    if n == 0:
        return 1
    return n * factorial_rec(n - 1)

factorial_rec(5)
```

## 3. Числа Фібоначчі (рекурсивно)
Асимптотична складність: **O(2^n)**

```

def fibonacci(n):
    if n == 0:
        return 0
    if n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

fibonacci(10)
```

## 4. Сортування злиттям
Асимптотична складність: **O(n log n)**

```

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def mergesort(arr):
    if len(arr) < 2:
        return arr
    mid = len(arr) // 2
    left = mergesort(arr[:mid])
    right = mergesort(arr[mid:])
    return merge(left, right)

mergesort([6, 5, 3, 1, 8, 7, 2, 4])
```


## Висновки
- Рекурсія дозволяє елегантно реалізувати алгоритми.
- Алгоритм Фібоначчі є неефективним без оптимізації.
- Merge Sort — ефективний алгоритм із стратегією «розділяй і володарюй».
